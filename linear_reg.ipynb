```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
plt.xlabel('Area (in sqr ft)')
plt.ylabel('Price (in US$)')
plt.scatter(df['area'], df['price'], color='red')   
```

```python
X = df[['area']]
y = df[['price']]
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# predict the house price when area=3300 sqr ft
lin_reg.predict([[3300]])
```

```python
f"Slope: {lin_reg.coef_}, Intercept: {lin_reg.intercept_}"
```

```python
# so the equation of our line is:
# y = 135.78767123*x + 135.78767123

# and our price prediction for the area=3300 matches the answer given by the equation of the line
135.78767123*3300 + 180616.43835616
```

```python
plt.xlabel('Area (in sqr ft)')
plt.ylabel('Price (in US$)')
plt.scatter(X, y, color='red') # make the scatter plot of the original points
plt.plot(X,lin_reg.predict(X)) # line of best fit according to linear regression
```
